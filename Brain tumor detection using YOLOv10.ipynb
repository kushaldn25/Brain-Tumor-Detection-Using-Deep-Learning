```python
!pip install -q roboflow ultralytics opencv-python gradio matplotlib

import os
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import torch
from ultralytics import YOLO
import gradio as gr
from pathlib import Path

# Set up GPU configuration
print(f"CUDA Available: {torch.cuda.is_available()}")
print(f"GPU Count: {torch.cuda.device_count()}")
if torch.cuda.is_available():
    for i in range(torch.cuda.device_count()):
        print(f"GPU {i}: {torch.cuda.get_device_name(i)}")

# Fix PyTorch 2.6+ security restrictions without recursion issues
# This approach avoids the recursion error by not patching torch.load directly
torch.hub._validate_not_a_forked_repo = lambda a, b, c: True
# Also set environment variable to allow weights_only=False
os.environ["TORCH_ALLOW_DANGEROUS_DESERIALIZATION"] = "1"

# Download and prepare dataset from Roboflow
from roboflow import Roboflow
rf = Roboflow(api_key="ljFPHCEea1M0E6rJYTEJ")
project = rf.workspace("brain-mri").project("mri-rskcu")
version = project.version(3)
dataset = version.download("yolov8")

# Get the dataset directory path
dataset_dir = os.path.join(os.getcwd(), "MRI-3")
print(f"Dataset downloaded to: {dataset_dir}")

# Verify dataset structure
print("\nDataset Structure:")
for split in ['train', 'valid', 'test']:
    path = os.path.join(dataset_dir, split, 'images')
    if os.path.exists(path):
        images = glob.glob(f"{path}/*.jpg")
        print(f"  {split}: {len(images)} images")

# Load the YOLO model
print("\nLoading base model...")
model_path = 'yolov8n.pt'
try:
    base_model = YOLO(model_path)
    print(f"Base model loaded successfully: {model_path}")
except Exception as e:
    print(f"Error loading model: {e}")
    # Try alternative download method if standard method fails
    import urllib.request
    print("Attempting alternative download method...")
    urllib.request.urlretrieve("https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt", model_path)
    base_model = YOLO(model_path)
    print(f"Base model loaded via alternative method: {model_path}")

# Create output directories
output_dir = os.path.join(os.getcwd(), "brain_tumor_detection")
os.makedirs(output_dir, exist_ok=True)
weights_dir = os.path.join(output_dir, "weights")
os.makedirs(weights_dir, exist_ok=True)
results_dir = os.path.join(output_dir, "results")
os.makedirs(results_dir, exist_ok=True)

# Train the model with GPU acceleration
print("\nStarting model training...")
results = base_model.train(
    data=os.path.join(dataset_dir, "data.yaml"),
    epochs=25,
    batch=16,
    imgsz=640,
    patience=5,
    device=0,  # Use first GPU
    project=output_dir,
    name="train",
    save=True,
    plots=True
)
print("Training completed")

# Plot training results
def plot_training_results():
    try:
        metrics_path = os.path.join(output_dir, "train", "results.csv")
        if os.path.exists(metrics_path):
            import pandas as pd
            metrics = pd.read_csv(metrics_path)
            
            plt.figure(figsize=(20, 12))
            
            # Plot training metrics
            plt.subplot(2, 2, 1)
            plt.plot(metrics['epoch'], metrics['train/box_loss'], label='train')
            plt.plot(metrics['epoch'], metrics['val/box_loss'], label='val')
            plt.title('Box Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            
            plt.subplot(2, 2, 2)
            plt.plot(metrics['epoch'], metrics['train/cls_loss'], label='train')
            plt.plot(metrics['epoch'], metrics['val/cls_loss'], label='val')
            plt.title('Class Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            
            plt.subplot(2, 2, 3)
            plt.plot(metrics['epoch'], metrics['train/dfl_loss'], label='train')
            plt.plot(metrics['epoch'], metrics['val/dfl_loss'], label='val')
            plt.title('DFL Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()
            
            plt.subplot(2, 2, 4)
            plt.plot(metrics['epoch'], metrics['metrics/precision(B)'], label='precision')
            plt.plot(metrics['epoch'], metrics['metrics/recall(B)'], label='recall')
            plt.plot(metrics['epoch'], metrics['metrics/mAP50(B)'], label='mAP50')
            plt.plot(metrics['epoch'], metrics['metrics/mAP50-95(B)'], label='mAP50-95')
            plt.title('Performance Metrics')
            plt.xlabel('Epoch')
            plt.ylabel('Metric Value')
            plt.legend()
            
            plt.tight_layout()
            plt.savefig(os.path.join(output_dir, 'training_metrics.png'))
            plt.show()
            print("Training metrics plotted")
        else:
            print(f"Training metrics file not found: {metrics_path}")
    except Exception as e:
        print(f"Error plotting training metrics: {e}")

# Plot training results
plot_training_results()

# Load the best model weights
best_weights_path = os.path.join(output_dir, "train", "weights", "best.pt")
if os.path.exists(best_weights_path):
    print(f"\nLoading best model weights from: {best_weights_path}")
    model = YOLO(best_weights_path)
    print("Best model weights loaded successfully")
else:
    print(f"Best weights not found at {best_weights_path}, using base model")
    model = base_model

# Validate the model - FIXED to handle numpy array formatting issues
def validate_model():
    valid_dir = os.path.join(dataset_dir, "valid", "images")
    if os.path.exists(valid_dir):
        print(f"\nRunning validation on {valid_dir}")
        val_results = model.val(data=os.path.join(dataset_dir, "data.yaml"), split="val")
        
        # Print validation metrics - Fix for numpy array formatting
        print("\nValidation Metrics:")
        print(f"mAP50: {float(val_results.box.map50):.4f}")
        print(f"mAP50-95: {float(val_results.box.map):.4f}")
        
        # Careful handling of potentially array values
        if hasattr(val_results.box, 'p'):
            p_value = val_results.box.p
            if isinstance(p_value, (np.ndarray, list)):
                print(f"Precision: {float(p_value[0] if len(p_value) > 0 else 0):.4f}")
            else:
                print(f"Precision: {float(p_value):.4f}")
        
        if hasattr(val_results.box, 'r'):
            r_value = val_results.box.r
            if isinstance(r_value, (np.ndarray, list)):
                print(f"Recall: {float(r_value[0] if len(r_value) > 0 else 0):.4f}")
            else:
                print(f"Recall: {float(r_value):.4f}")
        
        # Run predictions on validation set to generate visualization images
        pred_results = model.predict(
            source=valid_dir,
            conf=0.25,
            save=True,
            project=results_dir,
            name="validation_predictions"
        )
        print(f"Validation predictions completed: {len(pred_results)} images processed")
        return True
    else:
        print(f"Validation directory not found: {valid_dir}")
        return False

# Visualize detection results with side-by-side comparison
def visualize_detection_results(max_images=10):
    # Find prediction folder
    pred_dir = os.path.join(results_dir, "validation_predictions")
    if not os.path.exists(pred_dir):
        print(f"Prediction directory not found: {pred_dir}")
        return
    
    # Get all prediction images
    pred_images = sorted(glob.glob(os.path.join(pred_dir, "*.jpg")))
    if not pred_images:
        print(f"No prediction images found in {pred_dir}")
        return
    
    # Show side-by-side comparison (original vs prediction)
    images_to_show = pred_images[:max_images]
    n_images = len(images_to_show)
    
    if n_images == 0:
        print("No images to show")
        return
    
    # Create figure with two columns per image (original and prediction)
    fig, axes = plt.subplots(n_images, 2, figsize=(12, 4 * n_images))
    
    # Handle special case of single image
    if n_images == 1:
        axes = np.array([axes])
    
    for i, pred_path in enumerate(images_to_show):
        # Get original image path
        img_name = os.path.basename(pred_path)
        orig_path = os.path.join(dataset_dir, "valid", "images", img_name)
        
        # Show original image
        if os.path.exists(orig_path):
            try:
                orig_img = mpimg.imread(orig_path)
                axes[i, 0].imshow(orig_img)
                axes[i, 0].set_title(f"Original Image: {img_name}")
                axes[i, 0].axis('off')
            except Exception as e:
                print(f"Error loading original image {orig_path}: {e}")
                axes[i, 0].text(0.5, 0.5, f"Error loading image: {str(e)}", ha='center')
                axes[i, 0].axis('off')
        else:
            axes[i, 0].text(0.5, 0.5, "Original image not found", ha='center')
            axes[i, 0].axis('off')
        
        # Show prediction image
        try:
            pred_img = mpimg.imread(pred_path)
            axes[i, 1].imshow(pred_img)
            axes[i, 1].set_title(f"Prediction: {img_name}")
            axes[i, 1].axis('off')
        except Exception as e:
            print(f"Error loading prediction image {pred_path}: {e}")
            axes[i, 1].text(0.5, 0.5, f"Error loading image: {str(e)}", ha='center')
            axes[i, 1].axis('off')
    
    plt.tight_layout()
    plt.savefig(os.path.join(results_dir, 'validation_comparisons.png'), dpi=300)
    plt.show()
    print(f"Visualization saved to {os.path.join(results_dir, 'validation_comparisons.png')}")

# Run validation and show results
if validate_model():
    visualize_detection_results(max_images=5)

# Create a more advanced prediction function with visualization
def predict_and_visualize(image_path, conf_threshold=0.25):
    """
    Predicts tumor in brain MRI and visualizes the results
    
    Args:
        image_path: Path to the input image
        conf_threshold: Confidence threshold for detection
    
    Returns:
        Visualization figure
    """
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return None
    
    try:
        # Make prediction
        results = model.predict(source=image_path, conf=conf_threshold)
        
        if results and len(results) > 0:
            # Get result image with bounding boxes
            result = results[0]
            result_img = result.plot()
            result_img_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
            
            # Original image
            orig_img = cv2.imread(image_path)
            orig_img_rgb = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
            
            # Create figure for visualization
            fig, axes = plt.subplots(1, 2, figsize=(12, 6))
            
            # Plot original image
            axes[0].imshow(orig_img_rgb)
            axes[0].set_title("Original MRI Scan")
            axes[0].axis('off')
            
            # Plot prediction with annotations
            axes[1].imshow(result_img_rgb)
            
            # Add detection info as text
            if hasattr(result, 'boxes') and result.boxes is not None and len(result.boxes) > 0:
                boxes = result.boxes
                title = f"Detected {len(boxes)} tumor(s)"
                
                # Add more details about detections
                for i, box in enumerate(boxes):
                    conf = float(box.conf[0]) if hasattr(box.conf, '__len__') else float(box.conf)
                    cls = int(box.cls[0]) if hasattr(box.cls, '__len__') else int(box.cls)
                    cls_name = result.names[cls]
                    
                    # Safely extract bounding box coordinates
                    if hasattr(box, 'xyxy') and hasattr(box.xyxy, '__len__') and len(box.xyxy) > 0:
                        x1, y1 = float(box.xyxy[0][0]), float(box.xyxy[0][1])
                    else:
                        x1, y1 = 10, 10  # Default position if coordinates not available
                        
                    axes[1].text(
                        x1, y1 - 10,
                        f"{cls_name}: {conf:.2f}",
                        color='white', fontsize=10, 
                        bbox=dict(facecolor='red', alpha=0.7)
                    )
            else:
                title = "No tumors detected"
                
            axes[1].set_title(title)
            axes[1].axis('off')
            
            plt.tight_layout()
            return fig
        else:
            print("No results returned from model")
            return None
    except Exception as e:
        print(f"Error in prediction visualization: {e}")
        return None

# Test prediction on a sample image
def test_prediction():
    # Find a test image
    valid_images = glob.glob(os.path.join(dataset_dir, "valid", "images", "*.jpg"))
    if not valid_images:
        print("No validation images found")
        return
    
    # Select a sample image
    test_image = valid_images[0]
    print(f"\nTesting prediction on: {os.path.basename(test_image)}")
    
    try:
        # Run prediction and visualization
        fig = predict_and_visualize(test_image)
        if fig:
            plt.savefig(os.path.join(results_dir, 'sample_prediction.png'), dpi=300)
            plt.show()
            print(f"Sample prediction visualization saved to {os.path.join(results_dir, 'sample_prediction.png')}")
        else:
            print("Prediction visualization failed")
    except Exception as e:
        print(f"Error in test prediction: {e}")

# Run sample prediction
test_prediction()

# Create Gradio interface with advanced visualization and error handling
def gradio_predict(image):
    """Prediction function for Gradio interface"""
    try:
        # Save input image temporarily
        temp_input = os.path.join(results_dir, "temp_input.jpg")
        cv2.imwrite(temp_input, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
        
        # Process with model
        results = model.predict(source=temp_input, conf=0.25)
        
        if results and len(results) > 0:
            result = results[0]
            
            # Extract detection info
            detection_info = []
            if hasattr(result, 'boxes') and result.boxes is not None and len(result.boxes) > 0:
                for i, box in enumerate(result.boxes):
                    try:
                        # Safely extract values, handling both scalar and array types
                        conf = float(box.conf[0]) if hasattr(box.conf, '__len__') else float(box.conf)
                        cls = int(box.cls[0]) if hasattr(box.cls, '__len__') else int(box.cls)
                        cls_name = result.names[cls]
                        
                        # Safe extraction of bounding box coordinates
                        if hasattr(box, 'xyxy') and hasattr(box.xyxy, '__len__') and len(box.xyxy) > 0:
                            x1, y1, x2, y2 = map(int, box.xyxy[0])
                        else:
                            continue  # Skip this detection if coordinates not available
                            
                        detection_info.append({
                            "class": cls_name,
                            "confidence": f"{conf:.2f}",
                            "bbox": [x1, y1, x2, y2]
                        })
                    except Exception as e:
                        print(f"Error processing detection {i}: {e}")
                        continue
            
            # Get visualization
            result_img = result.plot()
            result_img_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
            
            # Clean up
            if os.path.exists(temp_input):
                try:
                    os.remove(temp_input)
                except Exception as e:
                    print(f"Error removing temp file: {e}")
                
            return result_img_rgb, detection_info if detection_info else "No tumors detected"
        else:
            return image, "No tumors detected"
    except Exception as e:
        print(f"Error in prediction: {e}")
        return image, f"Error: {str(e)}"

# Launch Gradio app with improved interface and error handling
def launch_gradio():
    try:
        with gr.Blocks(title="Brain Tumor Detection using YOLOv8") as app:
            gr.Markdown("# Brain Tumor Detection using YOLOv8")
            gr.Markdown("Upload an MRI brain scan to detect tumors. The model will identify and highlight tumor regions.")
            
            with gr.Row():
                with gr.Column():
                    input_image = gr.Image(label="Upload MRI Image")
                    submit_btn = gr.Button("Detect Tumors", variant="primary")
                
                with gr.Column():
                    output_image = gr.Image(label="Detection Results")
                    output_text = gr.Markdown(label="Detection Details")
            
            # Find example images
            examples = []
            example_dir = os.path.join(dataset_dir, "valid", "images")
            if os.path.exists(example_dir):
                example_files = glob.glob(os.path.join(example_dir, "*.jpg"))[:3]  # Limit to 3 examples
                if example_files:
                    examples = [[file] for file in example_files]
            
            if examples:
                gr.Examples(
                    examples=examples,
                    inputs=input_image
                )
            
            submit_btn.click(
                fn=gradio_predict,
                inputs=input_image,
                outputs=[output_image, output_text]
            )

            gr.Markdown("### Model Information")
            gr.Markdown(f"- Using YOLOv8 model trained on brain MRI dataset")
            gr.Markdown(f"- Training dataset: {os.path.basename(dataset_dir)}")
            gr.Markdown(f"- Model confidence threshold: 0.25")

        # Launch the app
        print("\nLaunching Gradio interface...")
        app.launch(share=True)
    except Exception as e:
        print(f"Error launching Gradio interface: {e}")
        print("Continuing with model export...")

# Launch Gradio interface
try:
    launch_gradio()
except Exception as e:
    print(f"Failed to launch Gradio interface: {e}")

# Export model for later use with error handling
try:
    export_path = os.path.join(weights_dir, "brain_tumor_model.onnx")
    print(f"\nExporting model to ONNX format: {export_path}")
    model.export(format="onnx", imgsz=640)
    print("Model export completed")
except Exception as e:
    print(f"Error exporting model: {e}")

# Function for external use with robust error handling
def predict_brain_tumor(image_path, conf_threshold=0.25, save_path=None):
    """
    Predicts brain tumor in MRI image with robust error handling
    
    Args:
        image_path: Path to input image
        conf_threshold: Detection confidence threshold
        save_path: Optional path to save visualization
    
    Returns:
        tuple: (detections list, output image with annotations)
    """
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        return [], None
    
    try:
        # Make prediction
        results = model.predict(source=image_path, conf=conf_threshold)
        
        detections = []
        output_image = None
        
        if results and len(results) > 0:
            result = results[0]
            
            # Get output image with bounding boxes
            output_image = result.plot()
            output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
            
            # Extract detection info with robust error handling
            if hasattr(result, 'boxes') and result.boxes is not None and len(result.boxes) > 0:
                for i, box in enumerate(result.boxes):
                    try:
                        # Safe extraction of values
                        conf = float(box.conf[0]) if hasattr(box.conf, '__len__') and len(box.conf) > 0 else float(box.conf)
                        cls = int(box.cls[0]) if hasattr(box.cls, '__len__') and len(box.cls) > 0 else int(box.cls)
                        cls_name = result.names[cls]
                        
                        # Safe extraction of bounding box
                        if hasattr(box, 'xyxy') and hasattr(box.xyxy, '__len__') and len(box.xyxy) > 0:
                            x1, y1, x2, y2 = map(float, box.xyxy[0])
                            
                            detections.append({
                                "class": cls_name,
                                "confidence": conf,
                                "bbox": [int(x1), int(y1), int(x2), int(y2)]
                            })
                    except Exception as e:
                        print(f"Error processing detection {i}: {e}")
                        continue
            
            # Save visualization if requested
            if save_path:
                try:
                    cv2.imwrite(save_path, cv2.cvtColor(output_image, cv2.COLOR_RGB2BGR))
                except Exception as e:
                    print(f"Error saving output image: {e}")
        
        return detections, output_image
    except Exception as e:
        print(f"Error in prediction: {e}")
        return [], None
```

```python
# Demo of the standalone prediction function
def demo_standalone_prediction():
    """Demo of standalone prediction function"""
    print("\n--- Standalone Prediction Function Demo ---")
    
    # Find a test image
    valid_images = glob.glob(os.path.join(dataset_dir, "valid", "images", "*.jpg"))
    if not valid_images:
        print("No test images available")
        return
    
    sample_image = valid_images[0]
    save_path = os.path.join(results_dir, "standalone_prediction.jpg")
    
    print(f"Running prediction on: {os.path.basename(sample_image)}")
    detections, _ = predict_brain_tumor(sample_image, conf_threshold=0.25, save_path=save_path)
    
    # Print results
    if detections:
        print(f"Found {len(detections)} detections:")
        for i, det in enumerate(detections):
            print(f"  {i+1}. {det['class']} (Confidence: {det['confidence']:.2f})")
        print(f"Result image saved to: {save_path}")
    else:
        print("No detections found")

# Run standalone prediction demo
demo_standalone_prediction()

print("\nPrediction function 'predict_brain_tumor()' is ready to use")
print("Example usage: detections, image = predict_brain_tumor('path/to/mri.jpg')")
print("Brain Tumor Detection System setup complete!")
```
